```python
# Jupyter-specific setup
%matplotlib inline

from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path

# === CONFIG ===
model_path = r"C:\Mohamed-2\crack_project\yolo11n_crack_custom6\weights\best.pt"
input_folder = Path(r"C:\Mohamed\real life images")
output_folder = Path(r"C:\Mohamed\crack_detection_outputs")
output_folder.mkdir(parents=True, exist_ok=True)

# === Load model ===
model = YOLO(model_path)

# === Supported image extensions ===
image_extensions = [".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff"]

# === Loop through images ===
for image_path in input_folder.glob("*"):
    if image_path.suffix.lower() not in image_extensions:
        continue

    # --- Load image ---
    img_bgr = cv2.imread(str(image_path))
    if img_bgr is None:
        print(f"❌ Failed to read {image_path}")
        continue
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

    # --- Run inference ---
    results = model(image_path)
    result = results[0]

    # --- Extract masks ---
    masks = result.masks.data.cpu().numpy() if result.masks else []
    overlay_img = img_rgb.copy()

    for mask in masks:
        mask_resized = cv2.resize(mask, (overlay_img.shape[1], overlay_img.shape[0]), interpolation=cv2.INTER_NEAREST)
        binary_mask = (mask_resized > 0.5).astype(np.uint8)

        red_overlay = np.zeros_like(overlay_img)
        red_overlay[:, :, 0] = 255  # Red

        # Apply transparent red overlay
        overlay_img = np.where(binary_mask[:, :, None], cv2.addWeighted(overlay_img, 0.6, red_overlay, 0.4, 0), overlay_img)

    # --- Convert to BGR for saving ---
    output_bgr = cv2.cvtColor(overlay_img, cv2.COLOR_RGB2BGR)

    # --- Save output image ---
    output_path = output_folder / image_path.name
    cv2.imwrite(str(output_path), output_bgr)
    print(f"✅ Saved: {output_path}")

print("🎉 All images processed and saved.")
```
